```python
import numpy as np
import cv2
from  matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat.png') 
cat = cv2.resize(img,(628,1024))  
# cat = cv2.cvtColor(cat, cv2.COLOR_RGB2BGR)
img = cv2.imread('tiger.png') 
tiger = cv2.resize(img,(628,1024))  
# tiger = cv2.cvtColor(tiger, cv2.COLOR_RGB2BGR)

```

```python
cat = cat.astype(np.uint8)
tiger = tiger.astype(np.uint8)
# cv2.imwrite('tiger2.png',tiger)
for i in range(cat.shape[0]):
    for j in range(cat.shape[1]):
        cat[i,j] = (cat[i,j]*0.5+tiger[i,j]*0.5).astype(np.uint8)
```

```python
cv2.imwrite('res.png',cat)
```

```python
## RGBarraynumpyndarraynumpy

img = cv2.imread('cat.png') 
cat = cv2.resize(img,(628,1024))  
cat = cv2.cvtColor(cat, cv2.COLOR_RGB2BGR)
img = cv2.imread('tiger.png') 
tiger = cv2.resize(img,(628,1024))  
tiger = cv2.cvtColor(tiger, cv2.COLOR_RGB2BGR)

## size

for i in range(11):
    res = i/10*cat + (10-i)/10*tiger
    # print(type(res))
    res = res.astype(np.uint8)
    cv2.imwrite(f'./output/res_img_{i}_{10-i}.png',cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
    plt.imshow(res)
    plt.show()
```
