```python
import matplotlib.pyplot as plt

def getValue(x1,x2,t):   ##如果直接看n次Bezier曲线的定义式可能看不明白为什么要定义这个函数，这里需要理解的是n次Bezier曲线的公式推导本质上是一个递推的过程
    return x1+(x2-x1)*t

def Input(x,y):
    ## n+1个点确定n次Bezier曲线(三点不共线的情况下)
    pointNum = int(input("请输入坐标的个数:"))
    for i in range(pointNum):
        x.append(int(input("请输入第" + str(i + 1) + "个坐标的x值:")))
        y.append(int(input("请输入第" + str(i + 1) + "个坐标的y值:")))
    return

def BezierPoint():
    for i in range(len(xLast)):
        plt.plot(xLast[i],yLast[i],'*')
    return
def Draw(x,y,num):
    if num==3:
        x0 = getValue(x[0], x[1], i / cnt)
        y0 = getValue(y[0], y[1], i / cnt)
        x1 = getValue(x[1], x[2], i / cnt)
        y1 = getValue(y[1], y[2], i / cnt)
        ##根据递推式子计算出Bezier上的当前点下一个点的坐标
        xLast.append(getValue(x0, x1, i / cnt))
        yLast.append(getValue(y0, y1, i / cnt))
        plt.plot(xLast[i], yLast[i], '*')
    else:
        xNext=[]
        yNext=[]
        #存放num=num-1时的控制点坐标值
        for j in range(num-1):
            ##进行递归迭代
            xNext.append(getValue(x[j], x[j + 1], i / cnt))
            yNext.append(getValue(y[j], y[j + 1], i / cnt))
        Draw(xNext,yNext,num-1)   ##函数会递归调用知道num = 3
    return

def Show():
    plt.grid(True)
    plt.plot(x, y)
    BezierPoint()
    return

#主函数
x=[]
y=[]
xLast=[]
yLast=[]
cnt = 10  ##我们包括端点设有11个Bezier曲线上的点
Input(x,y)
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(cnt+1):
    Draw(x,y,len(x))
plt.cla()
Show()
plt.plot(xLast,yLast)
plt.show()

## 输入0，0 1，5 2，7 3，4
```
